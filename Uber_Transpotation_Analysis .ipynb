## Data Cleaning (Handling Missing Values)

```python
uber.isnull().sum()
```

```python
uber.drop_duplicates()
```

## Data Manipulation ( Filtering Row By COndition)

```python
filtered_uber = uber[uber["MILES"] > 10]
filtered_uber
```

```python
## Group data calculate statistics 


Average_miles_By_Category = uber.groupby("CATEGORY")["MILES"].mean()
Average_miles_By_Category
```

```python
## Sorting the Data 

sorted_uber = uber.sort_values(by = "START_DATE")

sorted_uber
```

```python
##
uber["PURPOSE"].value_counts().plot.bar()
plt.xlabel("PURPOSE")
plt.ylabel("Count")
plt.title("Count of Trips by Purpose")
plt.show()
```

```python
uber["START_DATE"] = pd.to_datetime(uber["START_DATE"])
uber.set_index("START_DATE")["MILES"].plot.line()
plt.ylabel("Miles")
plt.title("Miles  Traveled over Time")
plt.show()
```

```python
sns.boxplot( x = 'MILES', y = 'PURPOSE', data=uber)
plt.xlabel("Purpose")
plt.ylabel("Miles")
plt.title("Box Plot : Miles By Purpose")
plt.show()
```
